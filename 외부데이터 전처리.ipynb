#### 고산

2021년 데이터는 시간대별로 되어 있지만 2022년 데이터는 분별로 데이터가 나타나져 있음

https://zzinnam.tistory.com/entry/Pandas-%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%A5%BC-%EA%B7%B8%EB%A3%B9%ED%99%94%ED%95%98%EB%8A%94-%EA%B0%80%EC%9E%A5-%EC%A2%8B%EC%9D%80-%EB%B0%A9%EB%B2%95groupby-Grouper

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('./제주도/고산_2022_01_분별.csv',encoding='cp949')
df2 = pd.read_csv('./제주도/고산_2022_02_분별.csv',encoding='cp949')
df3 = pd.read_csv('./제주도/고산_2022_03_분별.csv',encoding='cp949')
df4 = pd.read_csv('./제주도/고산_2022_04_분별.csv',encoding='cp949')
df5 = pd.read_csv('./제주도/고산_2022_05_분별.csv',encoding='cp949')
df6 = pd.read_csv('./제주도/고산_2022_06_분별.csv',encoding='cp949')
df7 = pd.read_csv('./제주도/고산_2022_07_분별.csv',encoding='cp949')
```

```python
print(df1.shape,df2.shape,df3.shape,df4.shape,df5.shape,df6.shape,df7.shape)
```

```python
df8 = pd.concat([df1,df2,df3,df4,df5,df6,df7])
df8 = df8.reset_index(drop=True)
df8
#df8.tail(2)
```

```python
# 1시간 간격
df8['일시'] = df8['일시'].astype(np.datetime64) #
dates = pd.date_range('2022-01-01', periods=5089, freq='1H')
df2 = pd.DataFrame(data=dates)
df2.rename(columns={0:'일시'},inplace=True)
df2['누적강수량(mm)'] = None
df2
df9 = df8[['일시','누적강수량(mm)']]
# merge
df3 = pd.merge(left=df2,right=df9,how='left',on='일시')
df3
df3 = df3[['일시','누적강수량(mm)_y']]
df3.rename(columns={'누적강수량(mm)_y':'누적강수량(mm)'},inplace=True)
df3
# 누적강수량 -> 강수량 -> 결측치 채우기
df3['누적강수량(mm)'] = df3['누적강수량(mm)'].diff()
df3.rename(columns={'누적강수량(mm)':'강수량(mm)'},inplace=True)
df3['강수량(mm)'] = df3['강수량(mm)'].fillna(method='bfill')
df3['강수량(mm)'] = df3['강수량(mm)'].clip(0)
df3
```

```python
df3.to_excel('./제주도/고산_2022_시간별_강수량.xlsx')
```

```python
def 기온 (df):
    df['일시'] = df['일시'].astype(np.datetime64) #
    new = df.groupby(pd.Grouper(key='일시',freq='h'))['기온(°C)'].mean().reset_index()
    new['기온(°C)'] = new['기온(°C)'].round(1) #
    return new
```

```python
def 풍속 (df):
    df['일시'] = df['일시'].astype(np.datetime64) #
    new = df.groupby(pd.Grouper(key='일시',freq='h'))['풍속(m/s)'].mean().reset_index()
    new['풍속(m/s)'] = new['풍속(m/s)'].round(1) #
    return new
```

```python
def 습도 (df):
    df['일시'] = df['일시'].astype(np.datetime64) #
    new = df.groupby(pd.Grouper(key='일시',freq='h'))['습도(%)'].mean().reset_index()
    new['습도(%)'] = new['습도(%)'].round(0) #
    return new
```

```python
result = 기온(df8)
result
```

```python
result = 풍속(df8)
result
```

```python
result = 습도(df8)
result
```

```python
result.to_excel('./제주도/고산_2022_시간별_기온.xlsx')
```

```python
result.to_excel('./제주도/고산_2022_시간별_풍속.xlsx')
```

```python
result.to_excel('./제주도/고산_2022_시간별_습도.xlsx')
```
