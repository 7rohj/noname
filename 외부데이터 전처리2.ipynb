### 고산

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('./제주도/고산_2021_시간별.csv',encoding='cp949')
df2 = pd.read_excel('./제주도/고산_2022_시간별_기온.xlsx',index_col=0)
df3 = pd.read_excel('./제주도/고산_2022_시간별_풍속.xlsx',index_col=0)
df4 = pd.read_excel('./제주도/고산_2022_시간별_습도.xlsx',index_col=0)
df5 = pd.read_excel('./제주도/고산_2022_시간별_강수량.xlsx',index_col=0)
```

```python
df1
```

#### 기온

```python
df1['일시'] = df1['일시'].astype(np.datetime64)
df1 = df1[['일시','기온(°C)']]
df1
result = pd.concat([df1,df2])
result = result.reset_index(drop=True)
result
```

```python
result.to_excel('./제주도/고산_기온_최종.xlsx')
```

#### 풍속

```python
df1['일시'] = df1['일시'].astype(np.datetime64)
df1 = df1[['일시','풍속(m/s)']]
df1
result = pd.concat([df1,df3])
result = result.reset_index(drop=True)
result
```

```python
result.to_excel('./제주도/고산_풍속_최종.xlsx')
```

#### 습도

```python
df1['일시'] = df1['일시'].astype(np.datetime64)
df1 = df1[['일시','습도(%)']]
df1
result = pd.concat([df1,df4])
result = result.reset_index(drop=True)
result
```

```python
result.to_excel('./제주도/고산_습도_최종.xlsx')
```

#### 강수량

```python
df1['일시'] = df1['일시'].astype(np.datetime64)
df1 = df1[['일시','강수량(mm)']]
df1['강수량(mm)'].fillna(0,inplace=True) # null값 대체
result = pd.concat([df1,df5])
result = result.reset_index(drop=True)
result
```

```python
result.to_excel('./제주도/고산_강수량_최종.xlsx')
```
